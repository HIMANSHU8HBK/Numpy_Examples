```python
import numpy as np

# Define the arrays
array_0d = np.array(7)
array_1d = np.array([1, 2, 3])
array_2d = np.array([[1, 2], [3, 4], [5, 6]])
array_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Function to format and print array attributes
def print_array_attributes(array, name):
    max_width = 50  # Set the maximum width for value display
    print(f"{name} Array:")
    print(f"{'-'*60}")
    print(f"{'Attribute':<25} {'Value':<35}")
    print(f"{'-'*60}")
    
    def format_value(value):
        return str(value).replace('\n', ' ').ljust(max_width)

    print(f"{'Example':<25} {format_value(array)}")
    print(f"{'ndim':<25} {array.ndim}")
    print(f"{'shape':<25} {array.shape}")
    print(f"{'size':<25} {array.size}")
    print(f"{'dtype':<25} {array.dtype}")
    print(f"{'itemsize':<25} {array.itemsize} bytes")
    print(f"{'nbytes':<25} {array.nbytes} bytes")
    print(f"{'strides':<25} {array.strides}")
    print(f"{'Transposed Array (T)':<25} {format_value(array.T)}")
    print(f"{'ndarray.flat':<25} {format_value(list(array.flat))}")
    print(f"{'-'*60}\n")

# Print attributes for each array
print_array_attributes(array_0d, "0D")
print_array_attributes(array_1d, "1D")
print_array_attributes(array_2d, "2D")
print_array_attributes(array_3d, "3D")
```
